```python
import torch
import torchvision
import torchvision.transforms as transforms
import pickle
import numpy as np
import matplotlib.pyplot as plt
import Augmentor
from PIL import Image
import cv2
```

```python
p = Augmentor.Pipeline()
p.flip_left_right(probability=0.6)
p.rotate(probability=0.6, max_left_rotation=9, max_right_rotation=9)
p.zoom_random(0.8, percentage_area=0.95)
#p.random_distortion(probability=1, grid_width=1, grid_height=1, magnitude=1)


transforms1 = torchvision.transforms.Compose([
    p.torch_transform(),
    torchvision.transforms.ToTensor()
    ])
```

```python
trainset = torchvision.datasets.ImageFolder(root='./CIFAR-10-images/train/', transform=transforms1)
trainloader = torch.utils.data.DataLoader(trainset, 
                                          batch_size=1,
                                          shuffle=True)
print(len(trainloader))

testset = torchvision.datasets.ImageFolder(root='./CIFAR-10-images/test/', transform=transforms1)
testloader = torch.utils.data.DataLoader(testset, 
                                         batch_size=1,
                                         shuffle=False, 
                                         num_workers=2)
print(len(testloader))

classes = ('airplane', 'car', 'bird', 'cat',
           'deer', 'dog', 'frog', 'horse', 'ship', 'truck')
```
